```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = np.random.randint(1,50,100)
```

```python
data
```

## Histograma

```python
plt.hist(data)
plt.show()
```

```python
plt.hist(data, bins=50, histtype='step')
plt.show()
```

## Boxplot

```python
plt.boxplot(data, vert=False, patch_artist=True)
plt.show()
```

```python
plt.boxplot(data, vert=False, patch_artist=True, notch=True)
plt.show()
```

```python
data = np.append(data,200)
plt.boxplot(data, vert=False, patch_artist=True, notch=True)
plt.show()
```

```python
data = np.append(data,200)
plt.boxplot(data, vert=False, patch_artist=True, notch=True, showfliers=True)
plt.show()
```

## Scatter

```python
N = 50
x = np.random.rand(N)
y = np.random.rand(N)
area = (30 * np.random.rand(N))**2
plt.scatter(x, y)
plt.show()
```

```python
N = 50
x = np.random.rand(N)
y = np.random.rand(N)
area = (30 * np.random.rand(N))**2
plt.scatter(x, y, s=area)
plt.show()
```

```python
N = 50
x = np.random.rand(N)
y = np.random.rand(N)
area = (30 * np.random.rand(N))**2
plt.scatter(x, y, s=area, alpha=0.5)
plt.show()
```

```python
N = 50
x = np.random.rand(N)
y = np.random.rand(N)
area = (30 * np.random.rand(N))**2
colors = plt.cm.viridis(np.random.rand(N))
plt.scatter(x, y, s=area, alpha=0.5, c=colors)
plt.colorbar()
plt.show()
```

```python
plt.scatter(x, y, s=area, alpha=0.5, c=colors, marker='p')
plt.colorbar()
plt.show()
```

```python
plt.scatter(x, y, s=area, alpha=0.5, c=colors, marker='x')
plt.colorbar()
plt.show()
```

```python
plt.scatter(x, y, s=area, alpha=0.5, c=colors, marker='o')
plt.colorbar()
plt.show()
```

```python
plt.scatter(x, y, s=area, alpha=0.5, c=colors, marker='<')
plt.colorbar()
plt.show()
```
