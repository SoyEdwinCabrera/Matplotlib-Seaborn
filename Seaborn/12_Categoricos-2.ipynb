```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.scatterplot(x='total_bill', y='tip', data=tips)
plt.show
```

```python
sns.scatterplot(x='total_bill', y='tip', data=tips, hue='day')
plt.show
```

```python
sns.scatterplot(x='total_bill', y='tip', data=tips, hue='day', style='time',size='size')
plt.legend(loc='center',bbox_to_anchor=(1.25,0.5))
plt.show
```

```python
markers = {'Lunch':'D', 'Dinner':'s'}
sns.scatterplot(x='total_bill', y='tip', data=tips, hue='day', style='time',size='size',markers=markers)
plt.legend(loc='center',bbox_to_anchor=(1.25,0.5))
plt.show
```

```python
fig = plt.figure(figsize=(8,8))
markers = {'Lunch':'D', 'Dinner':'s'}
sns.scatterplot(x='total_bill', y='tip', data=tips, hue='day', style='time',size='size',markers=markers)
plt.legend(loc='center',bbox_to_anchor=(1.25,0.5))
plt.show
```

```python
sns.lineplot(x='total_bill', y='tip', data=tips)
plt.show
```

```python
fig = plt.figure(figsize=(8,8))
markers = {'Lunch':'D', 'Dinner':'s'}
sns.lineplot(x='total_bill', y='tip', data=tips, hue='day', style='time',size='size',markers=markers)
plt.legend(loc='center',bbox_to_anchor=(1.25,0.5))
plt.show
```

```python
sns.relplot(data=tips,x='total_bill', y='tip', hue='day', style='time',size='size')
plt.show
```

```python
sns.relplot(data=tips,x='total_bill', y='tip', hue='day', style='time',size='size',col='time')
plt.show
```
