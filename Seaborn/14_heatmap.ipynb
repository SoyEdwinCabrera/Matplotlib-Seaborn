```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
tips.corr(numeric_only=True)
```

```python
sns.heatmap(tips.corr(numeric_only=True))
plt.show()
```

```python
sns.heatmap(tips.corr(numeric_only=True),annot=True)
plt.show()
```

```python
sns.heatmap(tips.corr(numeric_only=True),annot=True, cmap='coolwarm')
plt.show()
```

```python
sns.heatmap(tips.corr(numeric_only=True),annot=True, cmap='coolwarm', linewidths=5)
plt.show()
```

```python
sns.heatmap(tips.corr(numeric_only=True),annot=True, cmap='coolwarm', linewidths=5,linecolor='black')
plt.show()
```

```python
sns.heatmap(tips.corr(numeric_only=True),annot=True, cmap='coolwarm', linewidths=5,linecolor='black',vmin=0,vmax=1)
plt.show()
```

```python
sns.heatmap(tips.corr(numeric_only=True),annot=True, cmap='coolwarm', linewidths=5,linecolor='black',vmin=-1,vmax=1)
plt.show()
```

```python
sns.heatmap(tips.corr(numeric_only=True),annot=True, cmap='coolwarm', linewidths=5,linecolor='black',vmin=-1,vmax=1,cbar=True)
plt.show()
```

```python
sns.heatmap(tips.corr(numeric_only=True),annot=True, cmap='coolwarm', linewidths=5,linecolor='black',vmin=-1,vmax=1,cbar=False)
plt.show()
```
